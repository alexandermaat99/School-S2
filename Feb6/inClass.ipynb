```python
import pandas as pd
from sklearn.linear_model import LinearRegression


df= pd.read_csv('/Users/alexandermaat/Downloads/nba_salaries.csv')
df.head()

def basic_wrangling(df, missing_threshold=0.50, unique_threshold=0.95, messages=True):
    # features with too much missing data
    # features with too many unique values
    # features with one unique value
    import pandas as pd
    for col in df:
        missing = df[col].isna().sum() 
        unique  = df[col].nunique()
        rows = df.shape[0]

        if missing /rows >= missing_threshold:
            df.drop(columns=[col], inplace=True)
            if messages:
                print(f'Dropped {col} due to missing data')
        elif unique / rows >= unique_threshold:
            df.drop(columns=[col], inplace=True)
            if messages:
                print(f'Dropped {col} due to high uniqueness')
        elif unique == 1:
            df.drop(columns=[col], inplace=True)
            if messages:
                print(f'Dropped {col} due to no variance')
    return df

```

```python
basic_wrangling(df, messages=False)
```
